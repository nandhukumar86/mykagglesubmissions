```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```
df_train = pd.read_csv('titanic_train.csv')
df_test = pd.read_csv('titanic_test.csv')
```

```
df_train.head()
```

```
df_test.head()
```

```
PassengerId = df_test['PassengerId'].copy(deep=True)
```

```
PassengerId = pd.DataFrame(PassengerId, columns=['PassengerId'])
```

```
df_test.drop(columns= ['PassengerId'], inplace=True)
```

```
y_train = df_train['Survived'].copy(deep=True)
```

```
print(df_train.columns)
print(df_test.columns)
```

```
df_train.drop(columns= ['PassengerId', 'Survived'], inplace=True)
```

```
print(df_train.columns)
print(df_test.columns)
```

```
df = df_train
print('\nColumns:\n',df.columns)
print('\nData types:\n',df.dtypes)
print('\nSize:\n',df.size)
print('\nShape:\n',df.shape)
print('\nHead(5):\n',df.head())
print('\nTail(5):\n',df.tail())
print('\nNull checks:\n',df.isnull().sum())
print('\nDescribe:\n',df.describe().T)
```

```
df = df_test
print('\nColumns:\n',df.columns)
print('\nData types:\n',df.dtypes)
print('\nSize:\n',df.size)
print('\nShape:\n',df.shape)
print('\nHead(5):\n',df.head())
print('\nTail(5):\n',df.tail())
print('\nNull checks:\n',df.isnull().sum())
print('\nDescribe:\n',df.describe().T)
```

```
df_train['Age'].fillna(df_train.Age.median(), inplace=True)
df_test['Age'].fillna(df_test.Age.median(), inplace=True)
df_test['Fare'].fillna(df_test.Fare.median(), inplace=True)
```

```
df_train['Cabin'].fillna('X', inplace=True)
df_test['Cabin'].fillna('X', inplace=True)
```

```
df_train['Cabin'] = df_train['Cabin'].str[0]
df_test['Cabin'] = df_test['Cabin'].str[0]
```

```
df_train.head()
```

```
df_test.head()
```

```
df_train.drop(columns= ['Name','Ticket'], inplace=True)
df_test.drop(columns= ['Name','Ticket'], inplace=True)
```

```
df_train['Embarked'].fillna('X', inplace=True)
```

```
from sklearn.preprocessing import LabelEncoder

# Categorical to Continuous variables conversion 
df_train.loc[:,['Sex', 'Cabin', 'Embarked']] = df_train.loc[:,['Sex', 'Cabin', 'Embarked']].apply(LabelEncoder().fit_transform)
df_test.loc[:,['Sex', 'Cabin', 'Embarked']] = df_test.loc[:,['Sex', 'Cabin', 'Embarked']].apply(LabelEncoder().fit_transform)
```

```
df_train.head()
```

```
df_test.head()
```

```
sns.pairplot(df_train)
```

```
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
# should exclude categorical columns, but includes X,y.
cols = ['Age','Fare']
df_train[cols] = scaler.fit_transform(df_train[cols])
df_test[cols] = scaler.fit_transform(df_test[cols])
```

```
df_train.head()
```

```
df_test.head()
```

```
X_train = df_train
X_test = df_test
```

```
# from sklearn.neighbors import KNeighborsClassifier

# model = KNeighborsClassifier(n_neighbors = 31 , weights = 'distance') # k=>sqrt(n)

# from sklearn.linear_model import LogisticRegression

# model = LogisticRegression()

from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB

model = GaussianNB() # Features are Normal, Continuous

```

```
model.fit(X_train, y_train)
```

```
y_test_predict = model.predict(X_test)
# y_train_predict = model.predict(X_train)
```

```
y_test_predict
```

```
print('Training Score: ', model.score(X_train, y_train))
```

```
df = pd.concat([PassengerId, pd.DataFrame(y_test_predict)], axis=1, ignore_index=True)
df.columns = ['PassengerId','Survived']
```

```
df
```

```
df.to_csv('GaussianNB.csv')
```
